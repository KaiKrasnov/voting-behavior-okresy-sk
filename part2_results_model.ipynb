```python
from part2_results_model import *

db_path = "data\\converted_data.db"
```

```python
# urobme model bez testoveho okresu a pozrieme ako on predikuje testovy okres
results = Results(db_path)
test_okres = "Bratislava IV"

index_mask = (results.x.index == test_okres)

test_input = results.x.loc[index_mask].iloc[0]
test_output = pd.Series(
    results.w.loc[index_mask].values,
    index=results.y.loc[index_mask].values
)

results.x = results.x.loc[~index_mask]
results.y = results.y.loc[~index_mask]
results.w = results.w.loc[~index_mask]

model, coefs = results.classification()

pred = results.predict_one(model, test_input)

comparison = pd.DataFrame({
    f"predikcie modelu pre okres {test_okres}": pred,
    f"realny podiel hlasov v okrese {test_okres}": test_output
})
comparison
```

```python
# teraz urobme model bez testoveho okresu na zaklade iba dolezitych parametrov
# a pozrieme ako on predikuje testovy okres
impactful = results.find_most_impactful(coefs, top_k=10)

column_mask = (results.x.columns.isin(impactful.index))
impactful_test_input = test_input.loc[column_mask]

impactful_model, _ = results.classification(impactful.index)

impactful_pred = results.predict_one(impactful_model, impactful_test_input)

impactful_comparison = pd.DataFrame({
    f"predikcie modelu pre okres {test_okres}": impactful_pred,
    f"realny podiel hlasov v okrese {test_okres}": test_output
})
impactful_comparison
```

```python
# pozrieme na log-loss tych dvoch modelov
print("idealny log-loss:", results.logloss(test_output, test_output))
print("cely model:", results.logloss(test_output, pred))
print("model na iba dolezitych:", results.logloss(test_output, impactful_pred))
```

```python
# teraz natrenujeme model na vsetkych okresoch a pozrieme, ake parametre maju najvacsi priemerny vplyv;
# pre tieto parametre pozrieme politicke strany, ktore su najviac ovplyvnene tymto parametrom
full_results = Results(db_path)
full_model, coefs = full_results.classification()
impactful = full_results.find_most_impactful(coefs, 20)

impact = pd.DataFrame({
    "najpositivnejsi vplyv": coefs.loc[impactful.index].max(axis=1),
    "strana pre positivny vplyv": coefs.loc[impactful.index].idxmax(axis=1),
    "najnegativnejsi vplyv": coefs.loc[impactful.index].min(axis=1),
    "strana pre negativny vplyv": coefs.loc[impactful.index].idxmin(axis=1),
})
impact
```

```python
# teraz pozrieme, ake parametre najviac ovplyvnuju podiel hlasov pre jednotlive politicke strany
def top_features_with_coef(col: pd.Series, top_n: int = 5) -> list:
    return [
        f"{val:.4f} ({idx})"
        for idx, val in col.loc[col.abs().nlargest(top_n).index].items()
    ]

result = coefs.apply(top_features_with_coef)
result.T
```

```python
# pozrieme na to, ako zmena parametra okresa vplyva na podiely hlasov.
# na to pouzijeme model bez testovaceho okresu, zmenime testovaci vstup, a porovname vystupy

test_input_bigger_student_pct = test_input.copy()
test_input_bigger_student_pct["študent vysokej školy (%)"] *= 2
pred_bigger_student_pct = results.predict_one(model, test_input_bigger_student_pct)

test_input_less_student_pct = test_input.copy()
test_input_less_student_pct["študent vysokej školy (%)"] /= 2
pred_less_student_pct = results.predict_one(model, test_input_less_student_pct)

test_input_bigger_slovak_pct = test_input.copy()
test_input_bigger_slovak_pct["slovenská (%)"] *= 2
pred_bigger_slovak_pct = results.predict_one(model, test_input_bigger_slovak_pct)

test_input_less_slovak_pct = test_input.copy()
test_input_less_slovak_pct["slovenská (%)"] /= 2
pred_less_slovak_pct = results.predict_one(model, test_input_less_slovak_pct)


changed = pd.DataFrame({
    "nezmene": pred,
    "2x percento studentov": pred_bigger_student_pct,
    "0.5x percento studentov": pred_less_student_pct,
    "2x percento slovakov": pred_bigger_slovak_pct,
    "0.5x percento slovakov": pred_less_slovak_pct
})
changed
```
